```python
import json
import random
import os

from google.colab import drive

drive.mount('/content/drive')
```

```python
feverous_path = "/content/drive/Shareddrives/517 nlp project/data/FEVEROUS/feverous_train_challenges.jsonl"
fever_path = "/content/drive/Shareddrives/517 nlp project/data/FEVER/train.jsonl"
original_fever_path = "/content/drive/Shareddrives/517 nlp project/data/FEVER/original_dataset.jsonl"
```

```python
# Rename original FEVER train file
if os.path.exists(fever_path):
    os.rename(fever_path, original_fever_path)

# Output paths (same directory as original data)
feverous_output_dir = "/content/drive/Shareddrives/517 nlp project/data/FEVEROUS/"
fever_output_dir = "/content/drive/Shareddrives/517 nlp project/data/FEVER/"


feverous_splits = {
    "train": feverous_output_dir + "feverous_train.jsonl",
    "dev": feverous_output_dir + "feverous_dev.jsonl",
    "test": feverous_output_dir + "feverous_test.jsonl"
}
fever_splits = {
    "train": fever_output_dir + "fever_train.jsonl",
    "dev": fever_output_dir + "fever_dev.jsonl",
    "test": fever_output_dir + "fever_test.jsonl"
}

# Split ratios
train_ratio = 0.8
dev_ratio = 0.1
test_ratio = 0.1

def split_dataset(file_path, output_paths):
    """Splits a JSONL dataset into train, dev, and test sets."""
    with open(file_path, "r", encoding="utf-8") as f:
        data = [json.loads(line) for line in f]

    random.shuffle(data)
    total = len(data)
    train_split = int(total * train_ratio)
    dev_split = int(total * (train_ratio + dev_ratio))

    splits = {
        "train": data[:train_split],
        "dev": data[train_split:dev_split],
        "test": data[dev_split:]
    }

    for split, path in output_paths.items():
        with open(path, "w", encoding="utf-8") as f:
            for entry in splits[split]:
                f.write(json.dumps(entry) + "\n")
        print(f"Saved {split} set to {path} ({len(splits[split])} samples)")

# Run dataset splitting
split_dataset(feverous_path, feverous_splits)
split_dataset(original_fever_path, fever_splits)
```
