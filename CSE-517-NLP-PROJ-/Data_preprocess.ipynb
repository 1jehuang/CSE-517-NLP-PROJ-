```python
# Bamboogle
## For Bamboogle, train set + label = dev set = test set
## We format Bamboogle into [{'question': question, 'answer': answer}]
data = pd.read_csv("/content/drive/Shareddrives/517 nlp project/data/Bamboogle/Bamboogle.csv")
questions = data['Question'].tolist()
answers = data['Answer'].tolist()
processed_data = []
for i in range(len(questions)):
  q = questions[i]
  a = answers[i]
  if q is None or a is None or q == "" or a == "":
    continue
  processed_data.append({"question": questions[i], "answer": answers[i]})

# save the data to three json: train.json, dev.json, test.json
## save to the same folder, for this dataset, they are all the same
with open("/content/drive/Shareddrives/517 nlp project/data/Bamboogle/train.json", "w") as f:
  json.dump(processed_data, f, indent=4)
with open("/content/drive/Shareddrives/517 nlp project/data/Bamboogle/dev.json", "w") as f:
  json.dump(processed_data, f, indent=4)
with open("/content/drive/Shareddrives/517 nlp project/data/Bamboogle/test.json", "w") as f:
  json.dump(processed_data, f, indent=4)

processed_data[:10]
```

```python
# HotpotQA
## For HotpotQA, the test set doesn't have a ground truth answer
## We format HotpotQA into [
    # {'question': question,
    #  'answer': answer,
    #  'original_question': original_question,
    #  'original_context': original_context}],
## where the 'question' is the combination of context and question
## and the 'original_question' is the original question
## and the 'original_context' is the original context
import json
from tqdm import tqdm, trange
train_data = json.load(open("/content/drive/Shareddrives/517 nlp project/data/HotpotQA/hotpot_train_v1.1.json"))
dev_data = json.load(open("/content/drive/Shareddrives/517 nlp project/data/HotpotQA/hotpot_dev_fullwiki_v1.json"))
test_data = json.load(open("/content/drive/Shareddrives/517 nlp project/data/HotpotQA/hotpot_test_fullwiki_v1.json"))

processed_train_data = []
for data in tqdm(train_data):
  question = data['question']
  answer = data['answer']
  context = data['context']
  # concat the context into a string
  context_str = ""
  for c in context:
    context_str += c[0] + '\n'
    context_str += ' '.join(c[1]) + '\n'
  processed_question = context_str + '\n' + question
  processed_train_data.append({"question": processed_question,
                               "answer": answer,
                               "original_question": question,
                               "original_context": context_str})
processed_train_data[0]
# processed dev data
processed_dev_data = []
for data in tqdm(dev_data):
  question = data['question']
  answer = data['answer']
  context = data['context']
  # concat the context into a string
  context_str = ""
  for c in context:
    context_str += c[0] + '\n'
    context_str += ' '.join(c[1]) + '\n'
  processed_question = context_str + '\n' + question
  processed_dev_data.append({"question": processed_question,
                               "answer": answer,
                               "original_question": question,
                               "original_context": context_str})

processed_test_data = []
for data in tqdm(test_data):
  question = data['question']
  context = data['context']
  # concat the context into a string
  context_str = ""
  for c in context:
    context_str += c[0]
    context_str += ' '.join(c[1]) + '\n'
  processed_question = context_str + '\n' + question
  processed_test_data.append({"question": processed_question,
                               "original_question": question,
                               "original_context": context_str})
# save them into json with indent
with open("/content/drive/Shareddrives/517 nlp project/data/HotpotQA/train.json", "w") as f:
  json.dump(processed_train_data, f, indent=4)
with open("/content/drive/Shareddrives/517 nlp project/data/HotpotQA/dev.json", "w") as f:
  json.dump(processed_dev_data, f, indent=4)
with open("/content/drive/Shareddrives/517 nlp project/data/HotpotQA/test.json", "w") as f:
  json.dump(processed_test_data, f, indent=4)
```

```python
# SVAMP
## For SVAMP, train set + label = dev set = test set
## We format SVAMP into [{'question': question, 'answer': answer}]
## where question is the concatenation of Body and Question
'''
Example of original dataset:
{'ID': 'chal-1',
 'Body': 'Each pack of dvds costs 76 dollars. If there is a discount of 25 dollars on each pack',
 'Question': 'How much do you have to pay to buy each pack?',
 'Equation': '( 76.0 - 25.0 )',
 'Answer': 51.0,
 'Type': 'Subtraction'}
'''
import json
data = json.load(open("/content/drive/Shareddrives/517 nlp project/data/SVAMP/SVAMP.json"))
processed_data = []
for d in data:
  question = d['Body'] + '\n' + d['Question']
  answer = d['Answer']
  processed_data.append({"question": question, "answer": answer})

# randomly split into train-dev-test, 7-1-2
import random
random.shuffle(processed_data)
train_data = processed_data[:int(len(processed_data)*0.7)]
dev_data = processed_data[int(len(processed_data)*0.7):int(len(processed_data)*0.8)]
test_data = processed_data[int(len(processed_data)*0.8):]

# print the lengths
print(len(train_data))
print(len(dev_data))
print(len(test_data))

# save
with open("/content/drive/Shareddrives/517 nlp project/data/SVAMP/train.json", "w") as f:
  json.dump(train_data, f, indent=4)
with open("/content/drive/Shareddrives/517 nlp project/data/SVAMP/dev.json", "w") as f:
  json.dump(dev_data, f, indent=4)
with open("/content/drive/Shareddrives/517 nlp project/data/SVAMP/test.json", "w") as f:
  json.dump(test_data, f, indent=4)

processed_data[:3]
```
